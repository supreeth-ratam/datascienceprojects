```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
```

```python
# Importing Dataset
car_sales_df = pd.read_csv('dataset.csv')
car_sales_df
```

```python
car_sales_df = car_sales_df.drop(columns=['Name', 'Location'])
car_sales_df
```

```python
car_sales_df.describe()
```

```python
car_sales_df.isna().sum()
car_sales_df = car_sales_df.dropna()
car_sales_df = car_sales_df.reset_index(drop=True)

car_sales_df.head()
```

```python
import re

car_sales_df['Mileage'] = pd.Series([re.sub('[^.0-9]', '', 
    str(val)) for val in car_sales_df['Mileage']], index = car_sales_df.index)

car_sales_df['Engine'] = pd.Series([re.sub('[^.0-9]', '', 
    str(val)) for val in car_sales_df['Engine']], index = car_sales_df.index)

car_sales_df['Power'] = pd.Series([re.sub('[^.0-9]', '', 
    str(val)) for val in car_sales_df['Power']], index = car_sales_df.index)
```

```python
car_sales_df = car_sales_df.replace(r'^\s*$', np.nan, regex=True)
car_sales_df.isna().sum()
car_sales_df = car_sales_df.dropna()

car_sales_df = car_sales_df.reset_index(drop=True)
car_sales_df
```

```python
car_sales_df['Mileage'] = pd.Series([float(str(val)) 
    for val in car_sales_df['Mileage']], index = car_sales_df.index)
car_sales_df['Engine'] = pd.Series([float(str(val)) 
    for val in car_sales_df['Engine']], index = car_sales_df.index)
car_sales_df['Power'] = pd.Series([float(val) 
    for val in car_sales_df['Power']], index = car_sales_df.index)
```

```python
sns.pairplot(car_sales_df)
```

```python
print(car_sales_df['Fuel_Type'].unique())
car_sales_df['Fuel_Type'] = pd.Categorical(car_sales_df['Fuel_Type'])
dfFuel_Type = pd.get_dummies(car_sales_df['Fuel_Type'], prefix = 'Fuel_Type')
dfFuel_Type.head()
```

```python
print(car_sales_df['Transmission'].unique())
car_sales_df['Transmission'] = pd.Categorical(car_sales_df['Transmission'])
dfTransmission = pd.get_dummies(car_sales_df['Transmission'], prefix = 'Transmission')
dfTransmission.head()
```

```python
print(car_sales_df['Owner_Type'].unique())
car_sales_df['Owner_Type'] = pd.Categorical(car_sales_df['Owner_Type'])
dfOwner_Type = pd.get_dummies(car_sales_df['Owner_Type'], prefix = 'Owner_Type')
dfOwner_Type
```

```python
car_sales_df = pd.concat([car_sales_df, dfFuel_Type, dfTransmission, dfOwner_Type], axis=1)
car_sales_df = car_sales_df.drop(columns=['Owner_Type', 'Transmission', 'Fuel_Type'])
car_sales_df
```

```python
min_threshold , max_threshhold = car_sales_df.Price.quantile([0.001 , 0.999])
min_threshold , max_threshhold
```

```python
car_sales_df[car_sales_df.Price < min_threshold ]
```

```python
car_sales_df[car_sales_df.Price > max_threshhold ]
```

```python
car_sales_df = car_sales_df[(car_sales_df.Price > min_threshold)  & (car_sales_df.Price < max_threshhold) ]
car_sales_df= car_sales_df.reset_index(drop=True)
car_sales_df
```

```python
car_sales_df.describe()
```

```python
X = car_sales_df.drop([ 'Price'], axis=1)
X
```

```python
y = car_sales_df['Price']
y
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
scaler.data_max_
```

```python
scaler.data_min_
```

```python
y = y.values.reshape(-1,1)
y.shape
```

```python
y_scaled = scaler.fit_transform(y)
y_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_scaled, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = Sequential()
model.add(Dense(256,activation='relu', input_dim = 16))
model.add(Dense(128,activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(optimizer= 'adam', loss='mean_squared_error')

epochs_hist=model.fit(X_train, y_train, epochs=200, batch_size= 100, verbose=1, validation_split=0.2)
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model loss progression during Training/Validation')
plt.ylabel('Loss')
plt.xlabel('No. of Epochs')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
X_Testing = X_test
y_predict = model.predict(X_Testing)
y_predict.shape
```

```python
print('Expected Price =', y_predict[:,0])
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
print('Percentage of data that fits our model =' , r2_score(y_test, y_predict)*100 , "%")
```
